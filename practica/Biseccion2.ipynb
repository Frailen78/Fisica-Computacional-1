```julia
a = 1;
b = 2;
pseudo0 = 1e-6
f(x) = x^2 - 3

vector = zeros(15)
for i in 1:length(vector)
    vector[i] = 1 / 10^i
end

iter = 0
iteraciones = zeros(length(vector))

for j in 1:length(vector)
    if  f(a)*f(b) > 0
        print("No se encontro una raiz")
    else
        for i in 1:100
            
            global iter = iter + 1
            pm = (a + b)/2 
    
            if f(a) * f(pm) < 0
                b = pm
            else
                a = pm 
            end
            
            if abs(f(pm)) < vector[j]
                print(pm)
               
                break
            end
    
        
        end 
    end

    iteraciones[j] = iter
end
```

```julia
using Plots
plot(vector, iteraciones, xlabel="Tolerancias", ylabel="Iteraciones", title="Biseccion")
scatter!(vector, iteraciones)
```
