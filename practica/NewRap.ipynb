```julia
function newrap(f,fp, x0, tol)
    iter = 0
    for i in 1:100
        x0 = x0 - f(x0)/fp(x0)
        iter = iter + 1
        if abs(f(x0)) < tol
            break
        end
    end
    return x0, iter
end
```

```julia
f(x) = x^2 - 3
fp(x) = 2*x

tols = zeros(15)
x0s = zeros(length(tols))
iters = zeros(length(tols))

for i in 1:length(tols)
    tols[i] = 1/10^(i)
    x0s[i], iters[i] = newrap(f ,fp ,1 , 1/10^(i))
end
```

```julia
using Plots
scatter(tols, iters,xscale=:log10)
plot!(tols, iters, xlabel="Tolerancias",ylabel="Iteraciones", title="Newton Rapshon")
```
