```julia
a = -1;
b = 5;
tol = 1e-10
f(x) = x^2 - 3
```

```julia
function Biseccion(tol,a,b,f)
    if  f(a)*f(b) > 0
        print("No se encontro una raiz")
    else
        iter = 0
        for i in 1:100
    
            pm = (a + b)/2 
            iter += 1
            if f(a) * f(pm) < 0
                b = pm
            else
                a = pm 
            end
            
            if abs(f(pm)) < tol
                return pm, iter
                break
            end
        end 
    end
end
```

```julia
tols = zeros(15)
xs = zeros(length(tols))
iters = zeros(length(tols))

for i in 1:length(tols)
    tols[i] = 1/10^(i)
    xs[i], iters[i] = Biseccion(1/10^(i),a,b,f)
end
```

```julia
using Plots
scatter(tols, iters,xscale=:log10)
plot!(tols, iters, xlabel="Tolerancias",ylabel="Iteraciones", title="Biseccion")
```
